```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from razredi import *
from jarvis_march import grid_peel_jarvis
from graham_scan import grid_peel_graham
import numpy as np
```

```python
#ne poganjaj tega. traja ene 15 min
seznam = []
for i in range(20):
    for j in range(10):
        mreza = enakomerna_mreza(i,j)
        ovojnice = grid_peel_graham(mreza)
        st_ovojnic = (i, j, len(ovojnice[1]))
        seznam.append(st_ovojnic)
#print(seznam)
```

```python
x = np.linspace(0,20,10000)
y = np.linspace(0, 10, 10000)
graf = plt.figure()
ax = graf.gca(projection = "3d")
ax.plot(*zip(*seznam), label = "Število ovojnic")
ax.plot(x, y, (x*y) **(1/3), label = "kr neki")
#plt.plot(x,x **(4/3), label = "n ^ (4/3)")
plt.legend()

#plt.savefig("st_ovojnic.jpg",bbox_inches='tight', dpi=600)
```
