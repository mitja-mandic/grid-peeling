```python
def narisi(rezultat_grid_peel):
    slovar_ovojnic = rezultat_grid_peel[1]
    for i in range(len(slovar_ovojnic)):
        seznam_x = []
        seznam_y = []
        for tocka in slovar_ovojnic[i]:
            seznam_x.append(tocka.x)
            seznam_y.append(tocka.y)
        seznam_x += [seznam_x[0]]
        seznam_y += [seznam_y[0]]
        plt.plot(seznam_x,seznam_y,"-o")
```

```python
f = plt.figure()
#narisi(jarvis)
narisi(graham)
plt.axis('square')

#f.savefig("5x5_potencna.png", bbox_inches='tight', dpi=600)
```

```python
#ne poganjaj tega. traja ene 15 min
seznam = []
seznam1 = []
for i in range(150):
    ovojnice = grid_peel_graham_potencna(i,i)
    st_ovojnic = (i, len(ovojnice[1]))
    st = len(ovojnice[1])
    seznam.append(st_ovojnic)
    seznam1.append(st)
```

```python
def power_law(x, a, b):
    return a*np.power(x, b)
x = [i for i in range(150)]
#print(x)
x1 = np.linspace(0, 150, 10000)
graf = plt.figure()
plt.plot(*zip(*seznam), label = "Število ovojnic")
pars, cov = curve_fit(f=power_law, xdata=x, ydata=seznam1)
#print(pars)
plt.plot(x1, 1.2 * x1, label = "1,2 * n")
plt.plot(x1, x1, label = "n")
plt.ylim(0, 800)
plt.legend()

#plt.savefig("st_ovojnic_potencna.jpg",bbox_inches='tight', dpi=600)
```
